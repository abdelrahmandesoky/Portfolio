```python
import requests
from bs4 import BeautifulSoup
import pandas as pd 
```

```python
title=[]
company_name=[]
location=[]
description=[]
links=[]
page=[]
```

```python
page_no=0
while page_no<200:
    result=requests.get('https://wuzzuf.net/search/jobs/?a=hpb&q=&start={}'.format(page_no))
    src=result.content
    soup=BeautifulSoup(src,'lxml')
    containers=soup.find_all('div',{'class':'css-1gatmva e1v1l3u10'})
    len(containers)
    page_no+=1

    for i in range (len(containers)):
        page.append(page_no)
        title.append(containers[i].find('h2',{'class':'css-m604qf'}).text)
        company_name.append(containers[i].find('a',{'class':'css-17s97q8'}).text.replace('-',''))
        location.append(containers[i].find('span',{'class':'css-5wys0k'}).text)
        description.append(containers[i].find('div',{'class':'css-y4udm8'}).text.replace('Time','Time | ').replace('Â·','|'))
        x=containers[i].find('h2',{'class':'css-m604qf'})
        links.append('https://wuzzuf.net'+x.find("a").attrs['href'])
    
    print('Page switched ',page_no)
```

```python
dftest=pd.DataFrame({'Links':links})
dftest.tail()
```

```python
df=pd.DataFrame({'Page':page,'Title':title,'Company_name':company_name,'Address':location,'Description':description,'Links':links})
df.tail()
```
