```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
title=[]
type_building=[]
size=[]
location=[]
price=[]
page=[]
```

```python
page_no=0
while page_no<=200:
    result=requests.get('https://aqarmap.com.eg/en/for-sale/property-type/cairo/?page={}'.format(page_no))
    src=result.content
    soup=BeautifulSoup(src,'lxml')
    containers=soup.find_all('a',{"class":"card-details d-flex flex-column text-dark"})
    page_no+=1
    for i in range (len(containers)):
        title.append(containers[i].h2.text.replace('\n',''))
        type_building.append((containers[i].div.label.text.replace('\n','').strip()))
        a=containers[i].div.text.split()
        size.append((a[1]))
        location.append(containers[i].find('p',{"class":"card-details-address"}).text.replace('\n','').strip())
        price.append(containers[i].find('div',{"class":"price-container mb-4"}).text.replace('\n',''))
        page.append(page_no)
    print('Page Switched ',page_no)    
    
```

```python
new_size=[]
for i in range(len(size)):
    new_size.append(size[i].translate({ord(i): None for i in 'qwertyuiopasdfghjklzxcvbnmQWERTYUIOPASDFGHJKLZXCVBNM'}))
    
```

```python
df=pd.DataFrame({'Page':page,'Title':title,'Type':type_building,'Size':new_size,'Location':location,'Price':price})
```
